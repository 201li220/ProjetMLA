```python
from modified_transformers import BertConfig,BertForSequenceClassification
```

```python
config = BertConfig(adapter=True)
model = BertForSequenceClassification(config)
print(model)
```

```python
for param in model.parameters():
    param.requires_grad = False

for layer in model.bert.encoder.layer:
    for param in layer.attention.output.adapter.parameters():
        param.requires_grad = True
    for param in layer.output.adapter.parameters():
        param.requires_grad = True





        
```

```python
for name, param in model.named_parameters():
    print(f'{name}: requires_grad={param.requires_grad}')
```
